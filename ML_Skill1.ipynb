```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sales=pd.read_csv('supermarket_sales - Sheet1 (1).csv')
sales
```

```python
sales.head()
```

```python
sales.shape
```

```python
#no.of rows and columns
sales.columns
```

```python
sales.dtypes
```

```python
#time and date
sales['Date']=pd.to_datetime(sales['Date'])
sales['Date']
```

```python
sales.set_index('Date')
```

```python
sales.describe()
```

```python
sales.isnull().sum()
```

```python
avg_unit_price = sales["Unit price"].astype("float").mean(axis=0)
sales["Unit price"].replace(np.nan, avg_unit_price, inplace=True)
```

```python
sales["Unit price"]
```

```python
from scipy import stats 
mode = stats.mode(sales['Quantity'])
print(mode)
```

```python
sales['Quantity'].replace(np.nan, "mode", inplace=True)
```

```python
sales.dropna(inplace=True)
```

```python
sales.isnull().sum()
```

```python
np.round(sales.corr(),2)
```

```python
plt.figure(dpi=125)
sns.heatmap(np.round(sales.corr(),2),annot=True)
plt.show()
```

```python
plt.figure(figsize=(12,6),dpi=100)
sns.regplot(x='Tax 5%',y='gross income',data=sales,color='Red')
plt.xlabel('Tax 5%')
plt.ylabel('Gross Income')
plt.show()
```

```python
plt.figure(figsize=(12,6),dpi=100)
sns.regplot(x='Quantity',y='cogs',data=sales,color='green')
plt.xlabel('Quantity')
plt.ylabel('Cost of Goods Sale')
plt.title('Quantity v cost of Goods Sales',fontsize=15)
plt.show()
```

```python
#To see the distribution of diffierent ratings
plt.figure(dpi=125)
sns.distplot(sales['Rating'],kde=False)
plt.show()
```

```python
#To find Mean Rating
plt.figure(dpi=125)
sns.distplot(sales['Rating'],kde=False)
plt.axvline(x=np.mean(sales['Rating']),c='green',label='Mean Rating')
plt.legend()
plt.show()
```

```python
#To see the distribution of different ratings
plt.figure(dpi=125)
sns.histplot(sales['Rating'])
plt.show()
```

```python
sales.hist(figsize=(12,12))
plt.show()
```

```python
#Branch Count
plt.figure(dpi=125)
sns.countplot(sales['Branch'])
plt.xlabel('Branch Name')
plt.ylabel('Count')
plt.title('Which Branch is the most busy ?')
A,B,C = sales.Branch.value_counts()
print('Branch A -',A)
print('Branch B -',C)
print('Branch C -',B)
plt.show()
```

```python
plt.figure(dpi=125)
sns.countplot(sales['Gender'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of Gender')
A,B = sales.Gender.value_counts()

print('Male-',B)
print('Female-',A)
plt.show()
```

```python
plt.figure(dpi=125)
sns.countplot(y = 'Product line',hue = "Gender",data=sales)
plt.xlabel('Count')
plt.ylabel('Product Type')
plt.show()
```

```python
sns.catplot(x='Product line',y='Unit price',hue='City',data=sales,aspect=2,jitter=False)
plt.xlabel('Product Type')
plt.ylabel('Unit price')
plt.show()
```

```python
plt.figure(dpi=125)
sns.countplot(y = 'Product line',hue = "City",data=sales)
plt.xlabel('Count')
plt.ylabel('Product Type')
plt.show()
```

```python
plt.figure(dpi=125)
sns.countplot(y = 'Branch',hue = "Payment",data=sales)
plt.xlabel('Count')
plt.ylabel('Branch')
plt.show()
```

```python
plt.figure(dpi=125)
sns.countplot(y = 'City',hue = "Payment",data=sales)
plt.xlabel('Count')
plt.ylabel('Product line')
plt.show()
```

```python
plt.figure(dpi=125)
sns.countplot(sales['City'])
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Which City is most busy?')
A,B,C = sales.City.value_counts()

print('Yangon -',A)
print('Naypyitow-',C)
print('Mandalay-',B)
plt.show()
```
